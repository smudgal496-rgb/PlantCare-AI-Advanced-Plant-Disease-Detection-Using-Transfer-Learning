```python
# Install TensorFlow in the notebook environment if it's missing
%pip install tensorflow --quiet

import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Model

# 1) Define input shape and number of classes
IMG_HEIGHT = 224
IMG_WIDTH = 224
NUM_CLASSES = 38  # adjust for your plant disease dataset

# 2) Load MobileNetV2 base model with pretrained ImageNet weights
base_model = MobileNetV2(
    input_shape=(IMG_HEIGHT, IMG_WIDTH, 3),
    include_top=False,      # exclude MobileNet's default classification head
    weights='imagenet'      # load weights pre-trained on ImageNet
)

# 3) Freeze the base model so its weights are not updated during initial training
base_model.trainable = False

# 4) Add custom classification head on top
x = base_model.output
x = GlobalAveragePooling2D()(x)           # feature aggregation
x = Dense(128, activation='relu')(x)       # additional fully-connected layer
x = Dropout(0.5)(x)                        # optional dropout for regularization
outputs = Dense(NUM_CLASSES, activation='softmax')(x)

# 5) Build the complete model
model = Model(inputs=base_model.input, outputs=outputs)

# 6) Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 7) Model summary (optional)
model.summary()
```
